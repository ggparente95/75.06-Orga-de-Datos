# 2- Se tiene un registro de transacciones bancarias, de la forma (nro de transacción, tipo de transacción, cuenta origen, cuenta destino, fecha, hora, monto). Se pide resolver en Pandas:
A. Validar que todas las transacciones cuenten con un tipo de transacción.

B. Validar que para las transacciones del tipo transferencia, exista siempre tanto cuenta origen como cuenta destino.

C. Verificar que todas las transacciones del tipo transferencia, depósito y extracción cuenten con montos distintos de cero.

D. Indicar cuáles fueron las 10 transacciones de mayor monto.

E. Indicar cuál es el tipo de transacción que registra mayor monto promedio.

F. Indicar cuáles son las 5 cuentas con mayor cantidad de transacciones.

G. Indicar cuáles son las 5 cuentas con mayor monto involucrado.

H. Para el tipo de transacción con mayor cantidad de monto promedio, indicar cuales son
las 5 cuentas con más transacciones.

```python
l = [[100500,None,1,2,'20-10-2019','20:30',55000],
     [100501,None,2,1,'20-01-2019','20:30',25000],
     [100502,'deposito',4,3,'20-02-2019','20:30',65000],
     [100503,'transferencia',6,4,'20-03-2019','20:30',95000],
     [100504,'extraccion',4,3,'20-04-2019','20:30',25000],
     [100505,'extraccion',2,1,'20-05-2019','20:30',235000],
     [100506,'extraccion',7,3,'20-06-2019','20:30',15000],
     [100507,'deposito',4,3,'20-07-2019','20:30',115000],
     [100508,'deposito',6,1,'20-08-2019','20:30',85000],
     [100509,'extraccion',1,6,'20-09-2019','20:30',55000],
     [100510,'extraccion',4,2,'20-11-2019','20:30',55000],
     [100511,'transferencia',7,2,'20-12-2019','20:30',95000],
     [100512,'deposito',8,1,'20-07-2019','20:30',55000],
     [100513,'deposito',4,3,'20-03-2019','20:30',45000],
     [100514,'deposito',7,4,'20-02-2019','20:30',0],
     [100515,'transferencia',6,2,'20-01-2019','20:30',85000],
     [100517,'transferencia',None,6,'20-09-2019','20:30',0],
     [100518,'transferencia',1,7,'20-10-2019','20:30',545000],
     [100519,'transferencia',3,2,'20-12-2019','20:30',535000],
    ]
df = pd.DataFrame(l)
df.columns = ['nro_trans','tipo_trans','cta_origen','cta_dest','fecha','hora','monto']
df['fecha'] = pd.to_datetime(df['fecha'])
df
```

```python
# A. Validar que todas las transacciones cuenten con un tipo de transacción.
df = df.loc[~df['tipo_trans'].isnull(), :] 
```

```python
# B. Validar que para las transacciones del tipo transferencia, exista siempre tanto cuenta origen como cuenta destino.
ids_invalidos = df[(df['tipo_trans']=='transferencia') & ((df['cta_origen']).isnull() | (df['cta_dest'].isnull()))].index
drop = df.drop(index=ids_invalidos, inplace=True)
df
```

```python
# C. Verificar que todas las transacciones del tipo transferencia, depósito y extracción cuenten con montos distintos de cero.
ids_invalidos = df[(df['tipo_trans'].isin(['transferencia','extraccion','deposito'])) & (df['monto']<=0)].index
ids_invalidos
df.drop(index=ids_invalidos, inplace=True)
df
```

```python
# D. Indicar cuáles fueron las 10 transacciones de mayor monto.
top_10 = df.sort_values(by='monto',ascending=False)[0:10]
top_10
```

```python
# E. Indicar cuál es el tipo de transacción que registra mayor monto promedio.
tipo_trans = df.groupby('tipo_trans')['monto'].mean().idxmax()
print('El tipo de transaccion de mayor monto promedio es: ',tipo_trans)
```

```python
# F. Indicar cuáles son las 5 cuentas con mayor cantidad de transacciones.
cta_mas_trans = df.groupby('cta_origen').size().idxmax()
print('La cuenta con mayor cantidad de transacciones es la ',cta_mas_trans)
```

```python
# G. Indicar cuáles son las 5 cuentas con mayor monto involucrado.
monto_origen = df.groupby('cta_origen')['monto'].sum()
monto_destino = df.groupby('cta_dest')['monto'].sum()
final = monto_origen + monto_destino
cta_mas_monto = final.idxmax()
print('La cuenta con mayor monto involucrado es la ', cta_mas_monto)
```

```python
# H. Para el tipo de transacción con mayor cantidad de monto promedio, indicar cuales son las 5 cuentas con más transacciones.
df_trans_mas_monto = df[df['tipo_trans']==tipo_trans]
df_trans_mas_monto.groupby('cta_origen').size().sort_values(ascending=False).reset_index()[0:5]
```

# Dado dataframe: (cod_prod, cod_fabric, mes, vtas_mensuales)
A. (cod_prod, promedio_ventas)

B. (cod_prod, min_vtas_mensual
              max_vtas_mensual
              promedio_vtas_mensual)
              
C. (cod_prod, min_vtas_mensual, max_vtas_mensual, promedio_vtas_mensual)

D. (         , cod_prod1, cod_prod2, cod_prod3, ... 

    min_vtas
    max_vtas
    promedio_vtas                                    )
    
E. filas: todos los prod, columnas: todos los fabricantes, valor true o false

```python
# A.
l = [[100500,'Algebra',9,612],
     [100501,'Analisis',4,45],
     [100502,'Fisica',7,555],
     [100500,'Algebra',4,111],
     [100501,'Analisis',5,44],
     [100502,'Fisica',4,333],
     [100501,'Algebra',5,222],
     [100502,'Algebra',6,222],
     [100500,'Fisica',6,532],
     [100500,'Analisis',3,432],
     [100501,'Algebra',1,122],
     [100502,'Analisis',7,111],
     [100500,'Fisica',3,444],
     [100501,'Analisis',8,452],
     [100505,'Fisica',7,166],
     [100502,'Fabr40',7,233],
     [100503,'Fabr41',9,244],
     [100502,'Fabr40',10,444],
     [100502,'Fabr41',10,112],
     [100505,'Fabr42',11,443],
     [100505,'Fabr43',12,443],
    ]
df = pd.DataFrame(l)
df.columns = ['cod_prod','cod_fabr','mes','vtas_mensuales']
```

```python
# A. (cod_prod, promedio_ventas)
df.groupby(['cod_prod']).agg({'vtas_mensuales':'mean'}).reset_index().rename(columns={'vtas_mensuales':'promedio_ventas'})
```

```python
#B. (cod_prod, min_vtas_mensual 
#              max_vtas_mensual 
#              promedio_vtas_mensual)

new_df = df.groupby('cod_prod').agg(min_vtas=('vtas_mensuales','min'),\
                           max_vtas=('vtas_mensuales','max'),\
                           promedio_vtas_mensual=('vtas_mensuales','mean')\
                          ).reset_index()

pd.melt(new_df, id_vars=['cod_prod']).sort_values(by='cod_prod')
```
